```python
def findNonFullyTrypticPeptides(data):
    filteredData = pd.DataFrame(columns=data.columns)  # Initialize with same columns as input data
    trypticData = pd.DataFrame(columns=data.columns)
    for index, row in data.iterrows():
        peptide = row['pep_seq']
        prev_aa = row['pep_res_before']
        next_aa = row['pep_res_after']

        # Check if peptide starts after K or R
        n_term_tryptic = prev_aa in ['K', 'R']

        # Check if peptide ends with K or R and next AA is not P
        c_term_tryptic = peptide.endswith(('K', 'R')) and next_aa != 'P'

        # Determine peptide type
        if n_term_tryptic and c_term_tryptic:
            trypticData = pd.concat([trypticData, pd.DataFrame([row])], ignore_index=True)
        elif n_term_tryptic or c_term_tryptic:
            filteredData = pd.concat([filteredData, pd.DataFrame([row])], ignore_index=True)
        else:
            filteredData = pd.concat([filteredData, pd.DataFrame([row])], ignore_index=True)
    return filteredData
```

```python
# Checks for mods in FRAGPIPE MSFragger .TSV files
# cols: pep_var_mod | pep_start | pep_end
def checkForMods(data, *args, **kwargs):
    group = kwargs.get('group', None)
    protein = kwargs.get('protein', None)
    mod = kwargs.get('mod', None) 
    pepStart = kwargs.get('ProteinStart', None)
    pepEnd = kwargs.get('ProteinEnd', None)
    printRes = kwargs.get('printRes', True)
    modCount = 0
    
    if protein:
        data = data[data['prot_acc'].str.contains(protein, case=False, na=False)]
    
    oxidation_k = 'oxidation \\(k\\)'
    oxidation_P = 'oxidation \\(p\\)'
    oxidation_M = 'oxidation \\(m\\)'
    deamidation_NQ = 'deamidation \\(n\\)'
    galactosyly = 'galactosyly'
    glucosylgalactosyl = 'glucosylgalactosyl'
    
    if mod == 'oxidation (k)':
        mod = oxidation_K
    elif mod == 'oxidation (p)':
        mod = oxidation_P
    elif mod == 'oxidation (m)':
        mod = oxidation_M
    elif mod == 'deamidation':
        mod = deamidation_NQ
    elif mod == 'galactosyl':
        mod = galactosyly
    elif mod == 'glucosylgalactosyl': 
        mod = glucosylgalactosyl
    else:
        mod = mod
    
    # Finds the relevant data points within your range
    if pepStart and pepEnd:
        relevantRegions = data[data['pep_start'] >= pepStart]
        relevantRegions = relevantRegions[relevantRegions['pep_end'] <= pepEnd]
    else: 
        relevantRegions = data
    
    # Seperate out ',' seperated modifications in pep_var_mod column
    modsPresent = relevantRegions['pep_var_mod']
    modsPresent = list(modsPresent)
    fullModList = []
    
    for item in modsPresent:
        item = str(item)
        if ';' in item:
            fullModList.extend(item.split(';'))
        else: 
            fullModList.append(item)

    # Find number of individual modifications
    if mod:
        modCount = 0
        for s in fullModList:
            if mod in s:
                modCount += 1
        print('{} | {} found in region {}-{}'.format(mod, modCount, pepStart, pepEnd))
    else: 
        galactosylModCount = 0
        for s in fullModList:
            if 'Galactosyl (K)' in s:
                galactosylModCount += 1
        
        glucosylgalactosylModCount = 0
        for s in fullModList:
            if 'Glucosylgalactosyl (K)' in s:
                glucosylgalactosylModCount += 1
                
        oxidation_k_ModCount = 0
        for s in fullModList:
            if 'Oxidation (K)' in s:
                oxidation_k_ModCount += 1
                
        oxidation_p_ModCount = 0
        for s in fullModList:
            if 'Oxidation (P)' in s:
                oxidation_p_ModCount += 1
                
        oxidation_m_ModCount = 0
        for s in fullModList:
            if 'Oxidation (M)' in s:
                oxidation_m_ModCount += 1
                
        deamidation_ModCount = 0
        for s in fullModList:
            if 'Deamidated' in s:
                deamidation_ModCount += 1
        
        if kwargs.get('printRes' == True):
            if kwargs.get('mod'):
                print('{} | {} found in region aa{}-{}'.format(kwargs.get('mod'), modCount, pepStart, pepEnd))
                if group:
                    if pepstart:
                        relevantRegions.to_csv('./PTMs/{}_{}_{}_aa{}-{}_mods.csv'.format(group, protein, mod, pepStart, pepEnd))
                    else:
                        relevantRegions.to_csv('./PTMs/{}_{}_{}_mods.csv'.format(group, protein, mod))
            else:
                print(
                     '{} | {} found in region aa{}-{} | avg: {}'.format('Oxidation (K)', oxidation_k_ModCount, pepStart, pepEnd, round(oxidation_k_ModCount/3)),
                     '\n{} | {} found in region aa{}-{} | avg {}'.format('Oxidation (P)', oxidation_p_ModCount, pepStart, pepEnd, round(oxidation_p_ModCount/3)),
                     '\n{} | {} found in region aa{}-{} | avg {}'.format('Oxidation (M)', oxidation_m_ModCount, pepStart, pepEnd, round(oxidation_m_ModCount/3)),
                     '\n{} | {} found in region aa{}-{} | avg {}'.format('Deamidation (NQ)', deamidation_ModCount, pepStart, pepEnd, round(deamidation_ModCount/3)),
                     '\n{} | {} found in region {}-{} | avg {}'.format('Galactosyl (K)(178.0477)', galactosylModCount, pepStart, pepEnd, round(galactosylModCount/3)),
                     '\n{} | {} found in region {}-{} | avg {}'.format('Glucosylgalactosyl (K)(340.1006)', glucosylgalactosylModCount, pepStart, pepEnd, round(glucosylgalactosylModCount/3)))
                if group:
                    if pepStart:
                        relevantRegions.to_csv('./PTMs/{}_{}_aa{}-aa{}_mods.csv'.format(group, protein, pepStart, pepEnd))
                    else: 
                        relevantRegions.to_csv('./PTMs/{}_{}_mods.csv'.format(group, protein))
        ptm_avgs = {
            'data': data,
            'ptm_avgs': [round(oxidation_k_ModCount/3), round(oxidation_p_ModCount/3), round(oxidation_m_ModCount/3), round(deamidation_ModCount/3), round(galactosylModCount/3), round(glucosylgalactosylModCount/3)],
            'ptms': [oxidation_k_ModCount, oxidation_p_ModCount, oxidation_m_ModCount, deamidation_ModCount, galactosylModCount, glucosylgalactosylModCount]
        }
        return ptm_avgs
                 # '\n{} | {} found in region {}-{}'.format('Galactosyl (K)(178.0477)', galactosylModCount, pepStart, pepEnd),
                 # '\n{} | {} found in region {}-{}'.format('Glucosylgalactosyl (K)(340.1006)', glucosylgalactosylModCount, pepStart, pepEnd)
```

```python
def findAA(s, ptms, *args, **kwargs):
    regionStart = kwargs.get('regionStart', None) # peptide region start
    regionEnd = kwargs.get('regionEnd', None) # peptide region end
    
    listOfIndexes = []
    listOfAA = []
    listOfPositions = []
    listOfColours = []
    s = list(s)
    for m in ptms:
        
        aa = []
        mod = ''
        colour = ''
        
        # - What Amino-Acid should be highlighted, which colour should be assigned to it
        if 'Oxidation (P)' in m:
            aa.append('P')
            mod = 'ox'
            colour = '\033[91m'
        elif 'Oxidation (K)' in m:
            aa.append('K')
            mod = 'ox'
            colour = '\033[91m'
        elif 'Oxidation (M)' in m:
            aa.append('M')
            mod = 'ox'
            colour = '\033[91m'
        elif 'Deamidated (NQ)' in m:
            # We want to find the first occurrence of 'N' or 'Q' separately, then highlight the 2nd occurrence of either
            # depending on the one that appears first in the string.
            aa.append('N')
            aa.append('Q')
            mod = 'deamidation'
            colour = '\033[94m'
        elif 'galactosyl' in m:
            aa.append('K')
            mod = 'galactosyl'
            colour = '\033[92m'
        elif 'glucosylgalactosyl' in m:
            aa.append('K')
            mod = 'glucosylgalactosyl'
            colour = '\033[92m'
        else:
            mod = ''
        
        # - Find which aa to highlight
        modNumber = 1
        aaCounter = 1
        
        m = list(m)
        
        for let in m:
            try:
                float(let)  # Try to convert to float
                modNumber = float(let)
            except ValueError:
                modNumber = modNumber
        
        # - Find the index of the aa to highlight
        for i, a in enumerate(s):
            if a in aa:
                if aaCounter == modNumber:
                    listOfIndexes.append(i)
                    listOfAA.append(a)
                    listOfColours.append(colour)
                    if regionStart and regionEnd:
                        listOfPositions.append(regionStart + i - 1)
                    aaCounter = aaCounter + 1
                else:
                    aaCounter = aaCounter + 1
                
    collection = { 
        'aa': listOfAA,
        'indexes': listOfIndexes,
        'aaPosition': listOfPositions,
        'colours': listOfColours
    }
    
    return collection
```

```python
def coloriseString(s, ptms, indexes, colors):
    s = s[:50] + " " + s[50:100] + " " + s[100:]
    sString = list(s)
    
    for i, m in enumerate(ptms):

        # Add red color and reset color after each modified character
        if int(indexes[i]) < 50: 
            sString[int(indexes[i])] = colors[i] + sString[int(indexes[i])] + '\033[90m'
        elif int(indexes[i]) >= 50 and int(indexes[i]) < 100: 
            sString[int(indexes[i])+1] = colors[i] + sString[int(indexes[i])+1] + '\033[90m'
        else:
            sString[int(indexes[i])+2] = colors[i] + sString[int(indexes[i])+2] + '\033[90m'
    colorString = "".join(sString)
    print(colorString)
```

```python
def shapiroWilk(g1, g2):
    # Data for each group
    group1 = g1  # PTM counts for the first group
    group2 = g2  # PTM counts for the second group
    print(g1, g2)
    # Perform Shapiro-Wilk test on Group 1
    stat_group1, p_group1 = shapiro(group1)
    print(f'Group 1: Statistics={stat_group1}, p-value={p_group1}')

    # Perform Shapiro-Wilk test on Group 2
    stat_group2, p_group2 = shapiro(group2)
    print(f'Group 2: Statistics={stat_group2}, p-value={p_group2}')

    # Interpret the results
    alpha = 0.05
    if p_group1 > alpha:
        print('Group 1 looks normally distributed (fail to reject H0)')
    else:
        print('Group 1 does not look normally distributed (reject H0)')

    if p_group2 > alpha:
        print('Group 2 looks normally distributed (fail to reject H0)')
    else:
        print('Group 2 does not look normally distributed (reject H0)')

def statMW(g1, g2):
    # Data for each group
    group1 = g1  # PTM counts for the first group
    group2 = g2  # PTM counts for the second group

    # Perform the Mann-Whitney U Test
    MW_stat, MW_p_value = mannwhitneyu(group1, group2)

    # Display the result
    print(f'Mann-Whitney U Statistic: {MW_stat}, p-value: {MW_p_value}')

    # Interpret the result
    alpha = 0.05
    if MW_p_value > alpha:
        print('Fail to reject the null hypothesis - the distributions are similar')
    else:
        print('Reject the null hypothesis - the distributions are significantly different')

        
    # Perform Wilcoxon Rank-Sum Test (for independent samples)
    WC_stat, WC_p_value = stats.ranksums(group1, group2)

    print(f"Wilcoxon Rank-Sum Test statistic: {WC_stat}")
    print(f"p-value: {WC_p_value}")

    # Interpret the result
    alpha = 0.05
    if WC_p_value > alpha:
        print("Fail to reject the null hypothesis (no significant difference between the groups)")
    else:
        print("Reject the null hypothesis (significant difference between the groups)")
        
    return {
        'MW': [MW_stat, MW_p_value],
        'WC': [WC_stat, WC_p_value]
    }
    
```
